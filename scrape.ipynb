```python
#slector gadget for chrome add the extention

import requests
from bs4 import BeautifulSoup
import pandas as pd
   
```

```python
url = r'https://www.dsebd.org/latest_share_price_scroll_by_ltp.php'
source = requests.get(url)
source.ok
```

```python
soup = BeautifulSoup(source.content, 'lxml')
soup
```

```python
soup.title.string
```

```python
soup.link.attrs['href']
```

```python
quote = soup.find('table', attrs={"table table-bordered background-white shares-table fixedHeader"})
quote.thead.find_all('th')[1].string
quote
```

```python
heads = [item.string for item in quote.thead.find_all('th')]
print(heads)
```

```python
list=[]
for item in quote.find_all('td'):
        list.append(item.string)
print(list)
```

```python
nested_list = []
for i in range(0, len(list), 11):
        nested_list.append(list[i:i+11])
print(nested_list)
```

```python
#  col_name=['#', 'TRADING CODE', 'LTP*', 'HIGH', 'LOW', 'CLOSEP*', 'YCP*', 'CHANGE', 'TRADE', 'VALUE (mn)', 'VOLUME']
df = pd.DataFrame(nested_list, columns=['#', 'TRADING CODE', 'LTP*', 'HIGH', 'LOW', 'CLOSEP*', 'YCP*', 'CHANGE', 'TRADE', 'VALUE (mn)', 'VOLUME'])
df
```

```python
quote.find_all('a')[0].string
header=[]
for item in quote.find_all('a'):
        header.append(item.string)
print(header)
df['TRADING CODE']=header
df
```

```python
file_path = ''
file_name = 'data.csv'
df.to_csv(file_path + file_name, index=False)
```

```python
# Storing inside Db

import sqlite3 as sq3

with sq3.Connection('mydb.db') as conn:
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE DSEData1(
        SerialNo TEXT,
        TradingCode TEXT,
        LTP TEXT,
        HIGH TEXT,
        LOW TEXT,
        CLOSEP TEXT,
        YCP TEXT,
        CHANGE TEXT,
        TRADE TEXT,
        VALUE_mn TEXT,
        VOLUME TEXT
    )''')
```

```python
with sq3.Connection('mydb.db') as conn:
    cursor = conn.cursor()
    for row in nested_list:
        cursor.execute('''INSERT INTO DSEData VALUES(?,?,?,?,?,?,?,?,?,?,?)
        ''',(row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9],row[10]))
```
